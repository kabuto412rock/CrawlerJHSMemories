## Part1  第一個頁面取得每個圖片方塊的連結和連結內容的描述。

```python
driver.get('http://tw.class.uschoolnet.com/class/?csid=css000000107821&id=model8&cl=1273127126-8770-2448')
```

```python
# 第一個頁面取得每個圖片方塊的連結和連結內容的描述，並轉換成一個memories_sitelinks_list
row2_2 = driver.find_elements_by_class_name('jteacher_row2_2')
row2_1 = driver.find_elements_by_class_name('jteacher_row2_1')

def handle(row):
    memories_link_list = []
    for elem in row:
        a = (elem.find_element_by_tag_name('a'))
    #     a.click()
        link = a.get_attribute('href')
        texts = elem.text.splitlines()
        for t in texts:
            print('@', t)
        if len(texts) != 3:
            print("糟糕,",link, " 這個網址的描述不太對!將會跳過儲存")
            continue
        site = {"時戳": texts[0],"標題": texts[1], "描述": texts[2], "連結": link}

        memories_link_list.append(site)
    return memories_link_list

```

```python
a_list = handle(row2_1)
b_list = handle(row2_2)
print('開始合併a_list和b_list')
memories_link_list = copy.deepcopy(a_list) + copy.deepcopy(b_list)
print(memories_link_list)
print('全部完成，如果中途有錯會上面的過程出打印出來。')
```

## Part 2 取得所有圖片的下載連結，並將對應的連結塞到Part 1產生的memories_sitelinks_list

<!-- 圖片下載連結的樣式 -->

```python
def bestAttemp(findDocument):
    max_attemps = 10
    while True:
        try:
            a = findDocument()
            break
        except:
            time.sleep(0.5)
            max_attemps -= 1
        if max_attemps == 0:
            break
    return a

        
    return None
```

```python
# driver = webdriver.Chrome(executable_path="./chromedriver")

def handle(row):
    memories_link_list = []
    for elem in row:
        a = (elem.find_element_by_tag_name('a'))
    #     a.click()
        link = a.get_attribute('href')
        texts = elem.text.splitlines()
        for t in texts:
            print('@', t)
        if len(texts) != 3:
            print("糟糕,",link, " 這個網址的描述不太對!將會跳過儲存")
            continue
        site = {"時戳": texts[0],"標題": texts[1], "描述": texts[2], "連結": link}

        memories_link_list.append(site)
    return memories_link_list

def handle_images(s_link):
    print(s_link.get('連結'))
    driver.get(s_link.get('連結'))
    picture_count = int(s_link.get('描述')[1:-1])
    img_urls = []
    img_block = bestAttemp(lambda :driver.find_element_by_class_name('album_out_td'))
    time.sleep(0.55)
    img_block.click()
    time.sleep(0.55)
    for i in range(picture_count):
        print(img_block)
        time.sleep(1.5)
        img_tag = bestAttemp(lambda :driver.find_element_by_id('fullResImage'))
        img_url = bestAttemp(lambda : img_tag.get_attribute('src'))
        
        img_urls.append(img_url)
        print(img_url)
#       最後一個圖片沒有下一頁按鈕，直接返回 img_urls
        if i == picture_count-1:
            continue
            
        time.sleep(0.5)    
        next_btn = bestAttemp(lambda : driver.find_element_by_class_name('pp_next'))
        
        time.sleep(0.5)    
        bestAttemp(lambda : next_btn.click())
    print("完成 一頁Desu")
    return img_urls


def main():
    print('開始合併a_list和b_list')
    memories_link_list = copy.deepcopy(a_list) + copy.deepcopy(b_list)

    print(memories_link_list)
    print('全部完成，如果中途有錯會上面的過程出打印出來。')
    for i in range(len(memories_link_list)):
        memories_link_list[i]['圖檔urls'] = handle_images(memories_link_list[i])
    print('Super完成!!!!!!!')

    return memories_link_list
results = main()
```

## Part 4. 讀取JSON檔內的List，轉換成dict方便後續下載
### (若曾執行過Part 1~3並成功建立json檔，可從Part 4開始執行)

```python
# 1. 重新取得第一到第三階段 辛苦建立的對應圖片網址的字典檔(JSON格式)
results = []
with open(DOWNLOAD_LIST_JSON_NAME, 'r') as f:
    results = json.load(f)
```

```python
results
```

```python
def create_directory_if_not_existed(dir_name):
    if not os.path.exists(dir_name):
        os.mkdir(dir_name)
        print('建立資料夾:', dir_name)
    else:
        print("已經存在資料夾:"+dir_name)
```

```python
# 2. 建立一個download_dict方便後續執行
download_dict = {}
result_dir_name = 'result'.strip()
create_directory_if_not_existed(result_dir_name)

for result in results:
    article_directory_name = result_dir_name+'/'+result['時戳'].strip()+"_"+result['標題']
    
    # 除了建立字典之外，也新增之後圖片要存放的本地資料夾
    create_directory_if_not_existed(article_directory_name)
    
    article_directory_abs_path = os.path.abspath(article_directory_name)
    print(article_directory_abs_path)
    print()
    current_img_urls = result['圖檔urls']
    for current_img_url in current_img_urls:
        img_file_name = current_img_url.split('/')[-1]
        print('img_file_name:', img_file_name)
        img_file_path = article_directory_abs_path + '/' + img_file_name
        print('@url', current_img_url)
        print('$local', img_file_path)
        download_dict[current_img_url] = img_file_path
```

# 第五階段 使用第四階段的下載網址字典，下載到對應的資料夾＆名稱

```python
import multiprocessing as mp
import urllib.request
```

```python
# 下載圖片的函數
def downloa_image(internet_img_url, local_file_path):
    print(end='下載中...')
    urllib.request.urlretrieve(internet_img_url, filename=local_file_path)
    print('完成！')
```

```python
# 使用Mutliprocessing 
for img_url in download_dict.keys():
    local_file_path = download_dict[img_url]
    mp.Process(target=downloa_image, args=[img_url, local_file_path]).run()
    
```

```python
img_url_count = 0
for result in results:
    img_url_count += len(result['圖檔urls'])
    print('當前相簿\"'+result['標題']+'\"有'+str(len(result['圖檔urls']))+'個圖片連結')
print('所有相簿的圖片加總有', img_url_count,'張')
```
